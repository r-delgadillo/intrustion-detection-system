```python
import glob
import pandas as pd
import utils
from sklearn import preprocessing
import numpy as np
```

```python
# Get all the files in the specified directory
path_to_csv_files = r'D:\CIC'
file_paths = glob.glob(path_to_csv_files + "\*.csv")

# Print the number of files found 
print(f"Found {len(file_paths)} files")
```

```python
dfs = []
# Construct the pandas data frame into a single data frame
# Iterate through all the files and append the data frame to the array
for file in file_paths:
    df = utils.read_csv_chunked(file)
    dfs.append(df)
```

```python
# Concatenate all data into one data frame
df = pd.concat(dfs)

# Print the data frame columns
print('Number of rows and columns: ', df.shape)
```

```python
# Remove the following columns from the dataset
# Remove  'Dst IP', 'Flow ID', 'Src IP', 'Src Port', since it's only available in Thursday-20-02
# Remove 'Timestamp' since it's not useful to calculating if a packet is malicious

expected_features = [
    'ACK Flag Cnt',      'Active Max',        'Active Mean',       'Active Min',
    'Active Std',        'Bwd Blk Rate Avg',  'Bwd Byts/b Avg',    'Bwd Header Len',
    'Bwd IAT Max',       'Bwd IAT Mean',      'Bwd IAT Min',       'Bwd IAT Std',
    'Bwd IAT Tot',       'Bwd PSH Flags',     'Bwd Pkt Len Max',   'Bwd Pkt Len Mean',
    'Bwd Pkt Len Min',   'Bwd Pkt Len Std',   'Bwd Pkts/b Avg',    'Bwd Pkts/s',
    'Bwd Seg Size Avg',  'Bwd URG Flags',     'CWE Flag Count',    'Down/Up Ratio',
    'Dst Port',          'ECE Flag Cnt',      'FIN Flag Cnt',      'Flow Byts/s',
    'Flow Duration',     'Flow IAT Max',      'Flow IAT Mean',     'Flow IAT Min',
    'Flow IAT Std',      'Flow Pkts/s',       'Fwd Act Data Pkts', 'Fwd Blk Rate Avg',
    'Fwd Byts/b Avg',    'Fwd Header Len',    'Fwd IAT Max',       'Fwd IAT Mean',
    'Fwd IAT Min',       'Fwd IAT Std',       'Fwd IAT Tot',       'Fwd PSH Flags',
    'Fwd Pkt Len Max',   'Fwd Pkt Len Mean',  'Fwd Pkt Len Min',   'Fwd Pkt Len Std',
    'Fwd Pkts/b Avg',    'Fwd Pkts/s',        'Fwd Seg Size Avg',  'Fwd Seg Size Min',
    'Fwd URG Flags',     'Idle Max',          'Idle Mean',         'Idle Min',
    'Idle Std',          'Init Bwd Win Byts', 'Init Fwd Win Byts', 'Label',             
    'PSH Flag Cnt',      'Pkt Len Max',       'Pkt Len Mean',      'Pkt Len Min',       
    'Pkt Len Std',       'Pkt Len Var',       'Pkt Size Avg',      'Protocol',          
    'RST Flag Cnt',      'SYN Flag Cnt',      'Subflow Bwd Byts',  'Subflow Bwd Pkts',  
    'Subflow Fwd Byts',  'Subflow Fwd Pkts',  'Timestamp',         'Tot Bwd Pkts',      
    'Tot Fwd Pkts',      'TotLen Bwd Pkts',   'TotLen Fwd Pkts',   'URG Flag Cnt'
]

# We need to remove any unexpected features
# The "expected_features" array are features described on the
# CIC-IDS website any others should be removed.

# Array storing which features will be removed from the data frame
remove_columns = ['Timestamp']

# Iterate through all the columns in the data frame
# and add any unexpected features to "remove_columns" property
for feature in df.columns:
    if feature not in expected_features:
        remove_columns.append(feature)

# Remove the columns specified in "remove_columns"
print('Removing columns: ', remove_columns)
df = df.drop(remove_columns, axis=1)

# Print the data frame columns
print('Number of rows and columns: ', df.shape)
print(f'Remaining column list: \n{sorted(df.columns)}')
```

```python
# Display the label column values
print('Attack types')
print('---------------------')
print(df['Label'].value_counts())
```

```python
# We want to replace all the attack labels with a generic "Malcious" label for our classifier problem
# That way it will result to "0" or "1" for benign or malicous respectfully

# This might be the complete list
attack_types = [
    'Bot',
    'Brute Force -Web',
    'Brute Force -XSS',
    'DDOS attack-HOIC',
    'DDOS attack-LOIC-UDP',
    'DDoS attacks-LOIC-HTTP',
    'DoS attacks-GoldenEye',
    'DoS attacks-Hulk',
    'DoS attacks-SlowHTTPTest',
    'DoS attacks-Slowloris',
    'Infilteration', 
    'SQL Injection',
    'FTP-BruteForce',
    'SSH-Bruteforce',
    'Label'                        # Some of the row data contains "Label" as a value instead of an attack type. Assume, it's malicious
]

#data_df = data_df.replace(to_replace=replace_labels, value="Malicious")
df[['Label']] = df[['Label']].replace(to_replace=attack_types, value="Malicious")
```

```python
# Display the label column values
print('Classification labels')
print('---------------------')
print(df['Label'].value_counts())
```

```python
# Instanstiate label encoder for use
label_encoder = preprocessing.LabelEncoder()

# Encode all strings to numerical values
df['Label'] = label_encoder.fit_transform(df['Label'].values)

# Print result of encoding (Should match above)
df['Label'].value_counts()
```
